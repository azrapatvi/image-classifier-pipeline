```python
features = [
    "people walking",
    "holi",
    "accident crash"
]
```

```python
import requests

access_key = "004wtdR0XdOj7G7YsoFi76kBW9xRWAYkye5uQahdVNU"

headers = {
    "Authorization": f"Client-ID {access_key}"
}

url = "https://api.unsplash.com/search/photos"

counter=1
for feature in features:
    params={
        "query":feature,
        "per_page":30, # to download 30 images per feature
        "page":1
    }

    response=requests.get(url,headers=headers,params=params)

    data = response.json()

    for img in data['results']:
        img_url=img['urls']['small'] # this only fetches the image url 

        try:
            img_data=requests.get(img_url) # this is actually repsonsible for taking binary content of an image

            with open(f"downloads_temp/img{counter}.jpg",'wb')as f:
                f.write(img_data.content)
            counter+=1
        except Exception as e:
            print("Error:", e)

    print(f"\nDownloaded 30 images for: {feature}\n")
```
